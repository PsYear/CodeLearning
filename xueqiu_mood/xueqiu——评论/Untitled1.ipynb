```python
#coding:utf-8
import random
import matplotlib.pyplot as plt

list_plt = []
for i in range(5000):
        list_plt.append(random.uniform(0, 1))
        
plt.plot(sorted(list_plt,reverse=True),"+")
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False


plt.xlabel(u"the number of comment from XueQiu ")
plt.ylabel(u"score")
plt.title(u"The score based on the dictionary of mood")
plt.show()
```
